```python
# taking engine_hp, city_mpg, popularity
xi = [170, 25, 873]
```

```python
# Bia terms
w0 = 7.17
w = [0.01, 0.04, 0.002]
```

```python
# g(Xi) = W0 + E(0 to n-1) WjXij

def linear_regression(xi):
    n = len(xi)

    pred = w0

    for j in range (n):
        pred = pred + w[j] * xi[j]
    
    return pred
```

```python
# LINEAR REGRESSION VECTOR FORM
```

```python
# g(Xi) = W0 + XiT * W

def dot(xi, w):
    n = len(xi)

    res = 0.0

    for j in range(n):
        res = res + xi[j] * w[j]
    return res
```

```python
def linear_regression(xi):
        
    return w0 + dot(xi, w)
```

```python
w_new = [w0] + w
w_new
```

```python
def linear_regression(xi):
    xi = [1] + xi
        
    return dot(xi, w_new)
```
