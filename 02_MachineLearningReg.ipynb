```python
df = pd.read_csv('data.csv')
df
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
df.head()
```

```python
df.make = df.make.str.lower().str.replace(' ','_')
df.make
```

```python
strings = list(df.dtypes[df.dtypes == 'object'].index)
strings
```

```python
for cols in strings:
    df[cols] = df[cols].str.lower().str.replace(' ','_')
```
